```python
# Install any extra packages if needed (Colab already has most)
!pip install seaborn plotly --quiet

# Import standard libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
import random

# Set plot style
sns.set(style="whitegrid")
np.random.seed(42)
random.seed(42)
```

```python
# Parameters
num_microbes = 50

# Generate microbes with random traits
microbes = pd.DataFrame({
    "Microbe_ID": [f"Microbe_{i+1}" for i in range(num_microbes)],
    "Growth_Rate": np.round(np.random.uniform(0.5, 2.0, num_microbes), 2),  # 0.5–2.0 units/day
    "Stress_Tolerance": np.round(np.random.uniform(0, 1, num_microbes), 2), # 0–1 scale
    "Metabolite_Production": np.round(np.random.uniform(10, 100, num_microbes), 2), # arbitrary units
    "Temperature_Optimum": np.random.randint(20, 45, num_microbes),  # °C
    "pH_Optimum": np.round(np.random.uniform(5.0, 9.0, num_microbes), 1)
})

microbes.head()
```

```python
# Environment parameters
temperatures = [25, 30, 35, 40]
pH_levels = [6.0, 6.5, 7.0, 7.5, 8.0]

# Growth simulation function
def simulate_growth(microbe, temp, pH, days=10):
    # Growth factor based on deviation from optimum
    temp_factor = max(0, 1 - abs(temp - microbe["Temperature_Optimum"])/20)
    pH_factor = max(0, 1 - abs(pH - microbe["pH_Optimum"])/3)
    stress_factor = microbe["Stress_Tolerance"]
    # Simulate cumulative growth
    growth = microbe["Growth_Rate"] * temp_factor * pH_factor * stress_factor * days
    return growth

# Run simulation for all microbes in all environments
simulation_results = []

for _, microbe in microbes.iterrows():
    for temp in temperatures:
        for pH in pH_levels:
            growth = simulate_growth(microbe, temp, pH)
            simulation_results.append({
                "Microbe_ID": microbe["Microbe_ID"],
                "Temperature": temp,
                "pH": pH,
                "Simulated_Growth": round(growth, 2)
            })

simulation_df = pd.DataFrame(simulation_results)
simulation_df.head()
```

```python
# Simple optimization: find environment (temp + pH) that maximizes growth
optimized_results = simulation_df.loc[simulation_df.groupby("Microbe_ID")["Simulated_Growth"].idxmax()]
optimized_results = optimized_results.merge(microbes, on="Microbe_ID")
optimized_results = optimized_results.sort_values(by="Simulated_Growth", ascending=False)
optimized_results.head()
```

```python
# Top 10 microbes
top_microbes = optimized_results.head(10)

plt.figure(figsize=(12,6))
sns.barplot(x="Microbe_ID", y="Simulated_Growth", data=top_microbes, palette="viridis")
plt.title("Top 10 Optimized Microbes by Simulated Growth")
plt.ylabel("Simulated Growth (units)")
plt.xticks(rotation=45)
plt.show()
```

```python
# Interactive heatmap using Plotly for a selected microbe
selected_microbe = optimized_results.iloc[0]["Microbe_ID"]

# Pivot the data correctly using keyword arguments
heatmap_data = simulation_df[simulation_df["Microbe_ID"] == selected_microbe].pivot(
    index="pH",
    columns="Temperature",
    values="Simulated_Growth"
)

import plotly.express as px

fig = px.imshow(
    heatmap_data.values,
    labels=dict(x="Temperature (°C)", y="pH", color="Growth"),
    x=heatmap_data.columns,
    y=heatmap_data.index,
    text_auto=True,
    aspect="auto"
)
fig.update_layout(title=f"Growth Heatmap for {selected_microbe}")
fig.show()
```

```python
# Save optimized results to CSV
optimized_results.to_csv("optimized_microbes.csv", index=False)
print("Results saved as optimized_microbes.csv")
```
